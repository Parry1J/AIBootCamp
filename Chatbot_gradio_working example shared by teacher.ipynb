```python
import gradio as gr
import logging
from transformers import pipeline
import torch

logging.basicConfig(level="INFO")
logger = logging.getLogger(__name__)
device = "cuda" if torch.cuda.is_available() else "cpu"

chat_model = pipeline("text2text-generation", model="google/flan-t5-xl", device=device)
logger.info(f"Model loaded on {device}")

def respond(
    message,
    history,
    system_message,
    max_tokens,
    temperature,
    top_p
):
    prompt = f"{system_message}"
    # preprocess message
    prompt += f"User: {message}\n Assistant:"
    logger.info(f"Generating response to {message}")
    response = chat_model(prompt, max_new_tokens=max_tokens, temperature=temperature, top_p=top_p, do_sample=True)[0]["generated_text"]
    response = response.split("Assistant:")[-1].strip()
    yield response

# Setting up the Gradio ChatInterface
demo = gr.ChatInterface(
    respond,
    additional_inputs=[
        gr.Textbox(value="You are a friendly Chatbot.", label="System message"),
        gr.Slider(minimum=1, maximum=2048, value=512, step=1, label="Max new tokens"),
        gr.Slider(minimum=0.1, maximum=4.0, value=0.7, step=0.1, label="Temperature"),
        gr.Slider(
            minimum=0.1,
            maximum=1.0,
            value=0.95,
            step=0.05,
            label="Top-p (nucleus sampling)",
        ),
    ],
)

if __name__ == "__main__":
    demo.launch(debug=True, share=True)
```
